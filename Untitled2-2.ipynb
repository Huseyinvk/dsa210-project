```python
# 4) Correlation matrix + descriptive
num_cols = ["diet_score","sleep_score","sleep_hours","caffeine","hydration","screen"]

plt.figure(figsize=(6,5))
sns.heatmap(df[num_cols].corr(), annot=True, cmap="coolwarm", linewidths=.5)
plt.title("Correlation Matrix")
plt.tight_layout()
plt.show()

df[num_cols].describe().T

```

```python
# 5) Multiple scatter / hist / distribution plots

plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x="diet_score", y="sleep_score")
plt.title("Diet Score vs Sleep Quality")
plt.show()

plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x="sleep_hours", y="sleep_score")
plt.title("Sleep Duration vs Sleep Quality")
plt.show()

plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x="caffeine", y="sleep_score")
plt.title("Caffeine vs Sleep Quality")
plt.show()

df["sleep_score"].hist(bins=8)
plt.title("Sleep Quality Distribution")
plt.xlabel("Sleep Score")
plt.ylabel("Count")
plt.show()

df["diet_score"].hist(bins=8)
plt.title("Diet Score Distribution")
plt.xlabel("Diet Score")
plt.ylabel("Count")
plt.show()
```

```python
# 6) Hypothesis testing
from scipy.stats import pearsonr, ttest_ind

df["healthy_diet"] = (df["diet_score"] > 6).astype(int)

r, p = pearsonr(df["healthy_diet"], df["sleep_score"])
print("Correlation diet-category vs sleep score:", r, p)

healthy = df[df["healthy_diet"] == 1]["sleep_score"]
unhealthy = df[df["healthy_diet"] == 0]["sleep_score"]
t, p2 = ttest_ind(healthy, unhealthy, equal_var=False)
print("T-test healthy vs unhealthy sleep quality:", t, p2)
```
